### Trending WeekDay, Hour

```python
df_video["snapshot_weekday"] = df_video["snapshot_time"].dt.day_name()

weekday_order = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]

df_video["snapshot_weekday"] = pd.Categorical(
    df_video["snapshot_weekday"],
    categories=weekday_order,
    ordered=True
)

ct_weekday = (
    pd.crosstab(
        df_video["snapshot_weekday"],
        df_video["is_trending"],
        normalize="index"
    ) * 100
).reindex(weekday_order)

ct_weekday
```

```python
# lay % trending
trending_rate = ct_weekday[True]

avg_rate = trending_rate.mean()

# mau sac
colors = [
    "#E74C3C" if x >= avg_rate else "#F5B7B1"
    for x in trending_rate
]

# ve chart
plt.figure(figsize=(10, 5))

bars = plt.bar(
    trending_rate.index,
    trending_rate.values,
    color=colors
)

# hien thi %
for bar in bars:

    height = bar.get_height()

    if pd.notna(height):

        plt.text(
            bar.get_x() + bar.get_width() / 2,
            height + 0.3,
            f"{height:.1f}%",
            ha="center",
            fontsize=9
        )

# line trung binh
plt.axhline(
    avg_rate,
    color="#3498DB",
    linestyle="--",
    linewidth=1,
    label=f"Trung binh: {avg_rate:.1f}%"
)

plt.title(
    "Tỷ lệ Video Trending theo Ngày trong Tuần (Snashot Weekday)",
    fontsize=14,
    weight="bold"
)

plt.xlabel("Weekday")
plt.ylabel("Tỷ lệ trending (%)")

plt.grid(
    axis="y",
    linestyle="--",
    alpha=0.3
)

plt.legend()

plt.show()
```

### Tạo feature Engagement cho ML

Tạo df_ml

```python
# Bước 1: Sort theo snapshot_time, lấy row đầu tiên mỗi video
df_ml = df_video.sort_values("snapshot_time").groupby("video_id").first().reset_index()

# Bước 2: Tính engagement
df_ml["like_rate"] = df_ml["likes"] / df_ml["views"]
df_ml["comment_rate"] = df_ml["comments"] / df_ml["views"]

# Bước 3: Lọc data hợp lệ
df_ml = df_ml[
    (df_ml["views"] > 0) &
    (df_ml["likes"] <= df_ml["views"])
].copy()

print(f"Rows: {df_ml.shape[0]}")
print(f"Trending: {df_ml['is_trending'].sum()}")
print(f"Not trending: {(~df_ml['is_trending']).sum()}")
```

```python
# Category encoding
# Music categories (20, 10) có trending rate cao → flag riêng
df_ml["is_music"] = df_ml["category_id"].isin([10, 20]).astype(int)

print("Feature: is_music")
print(df_ml.groupby("is_music")["is_trending"].agg(["sum", "count", "mean"]))
```

```python
# Hour 10, 22, 9 có trending rate cao
peak_hours = [9, 10, 21, 22, 23]
df_ml["is_peak_hour"] = df_ml["publish_hour"].isin(peak_hours).astype(int)

print("Feature: is_peak_hour")
print(df_ml.groupby("is_peak_hour")["is_trending"].agg(["sum", "count", "mean"]))
```

```python
# Friday, Saturday trending rate cao
weekend_or_friday = [4, 5, 6]  # Friday=4, Saturday=5, Sunday=6
df_ml["is_weekend"] = df_ml["publish_weekday"].isin(weekend_or_friday).astype(int)

print("Feature: is_weekend")
print(df_ml.groupby("is_weekend")["is_trending"].agg(["sum", "count", "mean"]))
```

```python
# HD vs SD
df_ml["is_hd"] = (df_ml["definition"] == "hd").astype(int)

print("Feature: is_hd")
print(df_ml.groupby("is_hd")["is_trending"].agg(["sum", "count", "mean"]))
```
